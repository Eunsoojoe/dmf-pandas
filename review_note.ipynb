## 분산(var)과 표준편차(std)

````python
# 평균 구하기 
total_mean = df['total_bill'].values.mean()
print(total_mean)

# 분산 구하기
total = ((df['total_bill'].values - total_mean) ** 2).sum()
total_count = df['total_bill'].count()-1
my_var = total / total_count
my_var_2 = df['total_bill'].var()
print(my_var, my_var_2)

my_std = np.sqrt(my_var)
my_std2 = df['total_bill'].std()
print(my_std, my_std2)
````

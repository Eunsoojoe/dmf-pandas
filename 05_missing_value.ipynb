### fillna

```python
df = sns.load_dataset('titanic')
```

```python
df1 = df.copy()
```

```python
df1['age'] = df1['age'].fillna(100)
df1.tail()
```

```python
df1['deck'].fillna('C')
```

```python
# df1['deck'].fillna('No Data')
# => 카테고리타입은 카테고리에 없는 정보는 넣을 수 없음.
```

```python
df1['deck'] = df1['deck'].cat.add_categories('No Data')
df1['deck'].fillna('No Data')
```

```python
df1 = df.copy()
df1['age']
```

```python
mean_age = df1['age'].mean()
df1['age'].fillna(mean_age).tail()
```

```python
median_age = df1['age'].median()
df1['age'].fillna(median_age).tail()
```

```python
# 남자 승객의 결측치는 남자 승객의 평균으로, 여자 승객의 결측치는 여자 승객의 평균으로

c1 = df1['sex'] == 'male'                                    # '남자 승객'이라는 조건 만들기
male_mean = df1.loc[c1, 'age'].mean()                        # '남자 승객' 평균 
df1.loc[c1,'age'] = df1.loc[c1, 'age'].fillna(male_mean)     # 결측치에 평균값 채우기

c2 = df1['sex'] == 'female'
female_mean = df1.loc[c2, 'age'].mean()
df1.loc[c2,'age'] = df1.loc[c2, 'age'].fillna(female_mean)

df1['age'].mean()
# mean_age_f = df1['age'=='female'].mean()
```

```python
df1['deck'].mode()
```

```python
df1['deck'].fillna(df1['deck'].mode()[0])    # 0번 인덱스를 추가해줘야 비로소 값에 접근을 할 수 있음.
```
